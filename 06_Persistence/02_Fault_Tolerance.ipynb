```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from langgraph.checkpoint.memory import MemorySaver
import time

load_dotenv()

```

```python

class StepState(TypedDict):
    input: str
    step1: str
    step2: str
    step3: str
    step4: str

```

```python

def step1(state: StepState):
    print("Running step 1...")
    return {"step1": "Successfully Executed"}


def step2(state: StepState):
    print("Running step 2...")
    time.sleep(30)
    return {"step2": "Successfully Executed"}


def step3(state: StepState):
    print("Running step 3...")
    return {"step3": "Successfully Executed"}


def step4(state: StepState):
    print("Running step 4...")
    return {"step4": "Successfully Executed"}

```

```python

checkpointers = MemorySaver()

graph = StateGraph(StepState)
graph.add_node("step1", step1)
graph.add_node("step2", step2)
graph.add_node("step3", step3)
graph.add_node("step4", step4)

graph.add_edge(START, "step1")
graph.add_edge("step1", "step2")
graph.add_edge("step2", "step3")
graph.add_edge("step3", "step4")
graph.add_edge("step4", END)

workflow = graph.compile(checkpointer=checkpointers)
thread_id = "1"
config = {"configurable": {"thread_id": thread_id}}
```

```python

result = workflow.invoke({"input": "Hello"}, config=config)
print(result)
```

```python
list(workflow.get_state_history(config=config))
```

```python
result = workflow.invoke(None, config=config)
print(result)
```
